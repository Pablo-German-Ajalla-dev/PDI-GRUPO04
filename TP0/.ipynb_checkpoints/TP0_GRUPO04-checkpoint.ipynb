```python
import tkinter as tk
from tkinter import filedialog, messagebox
from PIL import Image, ImageTk
import matplotlib.pyplot as plt

# Crear ventana principal oculta (solo para usar filedialog)
root = tk.Tk()
root.withdraw()

# --- Seleccionar imagen ---
ruta = filedialog.askopenfilename(
    title="Selecciona una imagen",
    filetypes=[("Imágenes", "*.jpg;*.jpeg;*.png;*.bmp;*.gif")]
)

if not ruta:
    messagebox.showinfo("Info", "No seleccionaste ninguna imagen.")
else:
    # Abrir imagen
    imagen = Image.open(ruta).convert("RGB")

    # Mostrar imagen original
    plt.imshow(imagen)
    plt.title("Imagen Original")
    plt.axis("off")
    plt.show()

    # --- Modificar pixeles (invertir colores como ejemplo) ---
    pixeles = imagen.load()
    for x in range(imagen.size[0]):
        for y in range(imagen.size[1]):
            r, g, b = pixeles[x, y]
            pixeles[x, y] = (255-r, 255-g, 255-b)

    # Mostrar imagen modificada
    plt.imshow(imagen)
    plt.title("Imagen Modificada (Colores invertidos)")
    plt.axis("off")
    plt.show()

    # --- Guardar ---
    guardar_ruta = filedialog.asksaveasfilename(
        defaultextension=".png",
        filetypes=[("PNG", "*.png"), ("JPEG", "*.jpg"), ("BMP", "*.bmp")]
    )
    if guardar_ruta:
        imagen.save(guardar_ruta)
        messagebox.showinfo("Éxito", f"Imagen guardada en:\n{guardar_ruta}")

```
